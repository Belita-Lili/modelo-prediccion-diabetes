```python
# Cargar los datos
diab = pd.read_csv("DEFUN_DIAB_TS.csv")
```

```python
# Convertir fechas a formato datetime
diab['FECHA_OCURR'] = pd.to_datetime(diab['FECHA_OCURR'])
```

```python
# Filtrar datos necesarios y hacer copia segura
df_model = defun[['CAUSA_DEF', 'SEXO', 'EDAD', 'DEFUN_DIAB']].copy()

# Convertir SEXO a numérico: 1 (Hombre), 0 (Mujer)
df_model['SEXO'] = df_model['SEXO'].map({'Hombre': 1, 'Mujer': 0})

# Eliminar filas con valores nulos
df_model = df_model.dropna()
```

```python
df_model['CAUSA_DEF'] = df_model['CAUSA_DEF'].apply(lambda x: 'Diabetes' if 'diabetes' in str(x).lower() else 'Otra')
df_model['CAUSA_DEF'] = df_model['CAUSA_DEF'].map({'Diabetes': 1, 'Otra': 0})
```

```python
df_model = df_model.dropna(subset=['CAUSA_DEF'])  # eliminar solo si es nulo en esa columna
df_model['CAUSA_DEF'] = df_model['CAUSA_DEF'].astype(str)
```
